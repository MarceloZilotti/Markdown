### 🔬 Conceitos Fundamentais

#### Tipos de Análise de Dados

1. **Análise Descritiva** 📋
   - O que aconteceu?
   - Resumo dos dados históricos

2. **Análise Preditiva** 🔮
   - O que pode acontecer?
   - Modelagem e previsão

3. **Análise Prescritiva** 🎯
   - O que devemos fazer?
   - Recomendações baseadas em dados
